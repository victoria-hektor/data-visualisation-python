*   Pandas: Basic plotting from dataframes.

```python
# Install Plotly (if not already installed)
!pip install plotly

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## **Step 2: Generate Synthetic Data**

Create a synthetic dataset to use for visualisation:

```python
# Generate synthetic data
np.random.seed(42)
num_samples = 500
data = {
    'Feature1': np.random.normal(50, 15, num_samples),
    'Feature2': np.random.uniform(10, 100, num_samples),
    'Feature3': np.random.exponential(30, num_samples),
    'Category': np.random.choice(['A', 'B', 'C', 'D'], num_samples)
}

# Create a DataFrame
df = pd.DataFrame(data)
print(df.head())
```

## **Step 3: Visualisation with Matplotlib**

### 1. Histogram

```python
# Histogram
plt.hist(df['Feature1'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Feature1')
plt.xlabel('Feature1')
plt.ylabel('Frequency')
plt.show()
```

### 2. Scatter Plot

```python
# Scatter Plot
plt.scatter(df['Feature1'], df['Feature2'], c=df['Feature3'], cmap='viridis', alpha=0.7)
plt.colorbar(label='Feature3')
plt.title('Feature1 vs Feature2')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```

## **Step 4: Enhanced Plots with Seaborn**
### 1. Pairplot

```python
# Pairplot for all numerical features
sns.pairplot(df, hue='Category', palette='bright')
plt.show()
```

### 2. Boxplot

```python
# Boxplot for Feature3
sns.boxplot(data=df, x='Category', y='Feature3', palette='pastel')
plt.title('Feature3 Distribution Across Categories')
plt.show()
```

## **Step 5: Interactive Visualisations with Plotly**

### 1. Scatter Plot

```python
# Interactive Scatter Plot
fig = px.scatter(df, x='Feature1', y='Feature2', color='Category',
                 size='Feature3', title='Interactive Feature1 vs Feature2',
                 labels={'Feature1': 'Feature 1', 'Feature2': 'Feature 2'})
fig.show()
```

### 2. Bar Plot

```python
# Interactive Bar Plot
category_counts = df['Category'].value_counts().reset_index()
category_counts.columns = ['Category', 'Count']  # Rename columns for clarity

fig = px.bar(category_counts,
             x='Category', y='Count', color='Category',
             labels={'Category': 'Category', 'Count': 'Count'},
             title='Category Distribution')
fig.show()
```

## **Step 6: Experiment with Styling**
### Customising Colours and Shapes

```python
# Custom Scatter Plot with Matplotlib
colours = {'A': 'red', 'B': 'blue', 'C': 'green', 'D': 'purple'}
shapes = {'A': 'o', 'B': 's', 'C': 'D', 'D': '^'}

for cat in df['Category'].unique():
    subset = df[df['Category'] == cat]
    plt.scatter(subset['Feature1'], subset['Feature2'],
                c=colours[cat], marker=shapes[cat], label=cat, alpha=0.7)

plt.title('Custom Scatter Plot by Category')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.legend(title='Category')
plt.show()
```
